requires_grad=True is the most important part here — it tells PyTorch to track all operations on this tensor for automatic gradient computation (autograd).


out = x.pow(2).sum():
This line creates a computation graph and calculates a scalar value.

x.pow(2) → squares every element → [[1, 1], [1, 1]]
.sum() → adds all elements together → 1 + 1 + 1 + 1 = 4

So out is a scalar tensor containing the value 4.0.

out.backward():

This is the most important line for automatic differentiation.
It computes the gradient of out with respect to all tensors that have requires_grad=True (in this case, x).
PyTorch uses backpropagation (chain rule) to calculate these gradients

x.grad:

Accesses the gradient that was computed and stored in the .grad attribute of tensor x.

out is like the loss.

x is like the weights/parameters.

x.grad tells us: “If you want to decrease out, you should change each value in x by this amount (the gradient).

```python
x = torch. tensor([[1.0000, -1.], [1., 1.]], requires_grad=True)
out = x.pow(2).sum()
#It tells us the direction to update x
out.backward()
# This tensor shows how much the final output (out) will change if you slightly increase each value in x.
x.grad
```
