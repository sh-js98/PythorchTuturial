Create a 3x3 tensor with floating-point numbers filled with zeros.

```python
torch.tensor(np.array([[0.0, 0. ,0],
                       [0., 0.00, 0],
                       [0.00, 0.,0]]))
```

This is a better way of writing it.

```python
# Best way (Recommended)
tensor2 = torch.zeros((3, 3))

# Also acceptable
tensor2 = torch.tensor([[0., 0., 0.],
                        [0., 0., 0.],
                        [0., 0., 0.]])
```
